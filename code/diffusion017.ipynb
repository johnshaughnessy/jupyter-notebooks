This notebook uses the model I trained by running the dreambooth example in huggingface/diffusers. 

```python
from diffusers import DiffusionPipeline
import torch

pipeline = DiffusionPipeline.from_pretrained("johnshaughnessy/dreambooth-dog-toy-example", torch_dtype=torch.float16, use_safetensors=True).to("cuda")
image = pipeline("a photo of sks dog running in the meadow with sunflowers", num_inference_steps=50, guidance_scale=7.5).images[0]
image
```
